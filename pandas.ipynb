```python
import pandas as pd
import json
from pathlib import Path
import sqlite3
from itertools import chain

json_directory = "classroom_data"

json_files = sorted(Path(json_directory).glob("classroom_data_*.json"),reverse=True)

with open(json_files[0],"r",encoding="UTF-8") as f:
    data = json.load(f)
    

data = pd.json_normalize(data['courses'])
```

```python

teachers = pd.DataFrame()
for i in range(len(data["teachers"])):
    result = pd.json_normalize(data["teachers"][i][j] for j in range(len(data["teachers"][i])))
    teachers = pd.concat([teachers, result], ignore_index=True)
teachers.drop("profile.permissions",inplace=True,axis= 1)
teachers.drop(columns=["profile.id","profile.name.familyName","profile.name.givenName"],inplace=True,axis= 1)
teachers
```

```python
students = pd.DataFrame()
for i in range(len(data["students"])):
    result = pd.json_normalize(data["students"][i][j] for j in range(len(data["students"][i])))
    students = pd.concat([students, result], ignore_index=True)
students.drop("profile.permissions",inplace=True,axis= 1)
students.drop(columns=["profile.name.givenName","profile.name.familyName","profile.id"],inplace=True,axis= 1)

students
```

```python
assignments = pd.DataFrame()
for i in range(len(data["assignments"])):
    result = pd.json_normalize(data["assignments"][i][j] for j in range(len(data["assignments"][i])))
    assignments = pd.concat([assignments, result], ignore_index=True)
# assignments.drop("profile.permissions",inplace=True,axis= 1)
parts = ['dueDate.year','dueDate.month','dueDate.day','dueTime.hours','dueTime.minutes']
assignments[parts] = assignments[parts].apply(pd.to_numeric, errors='coerce')

# NaN time → 00:00 (keep missing dates as NaT)
assignments['dueTime.hours']   = assignments['dueTime.hours'].fillna(0)
assignments['dueTime.minutes'] = assignments['dueTime.minutes'].fillna(0)

# rename to what to_datetime expects
cols = {
    'dueDate.year':'year',
    'dueDate.month':'month',
    'dueDate.day':'day',
    'dueTime.hours':'hour',
    'dueTime.minutes':'minute'
}
assignments['due_at'] = pd.to_datetime(assignments[list(cols)].rename(columns=cols), errors='coerce')
assignments.drop(axis= 1 , inplace= True, columns=["updateTime","dueDate.year",'dueDate.month','dueDate.day','dueTime.hours','dueTime.minutes'])
assignments
```

```python

material = pd.DataFrame()
nested = [x for x in assignments["materials"] if isinstance(x, list)]
flat = list(chain.from_iterable(nested))
material = pd.json_normalize([x for x in flat if isinstance(x, dict)], sep=".")
assignments.drop("materials",inplace=True,axis= 1)
assignments = pd.concat([assignments,material], ignore_index=True)
material
```

```python

submissions = pd.DataFrame()
nested = [x for x in assignments["submissions"] if isinstance(x, list)]
flat = list(chain.from_iterable(nested))
submissions = pd.json_normalize([x for x in flat if isinstance(x, dict)], sep=".")
submissions

mask = (submissions['state'] == "TURNED_IN") | (submissions['state'] == "RETURNED")

submissions = submissions[mask]
submissions.drop(axis=1,inplace=True,columns=["submissionHistory","updateTime",])
submissions
```
